#Appstore and Google Play Analysis

- Our goal for this project is to analyze data to help our developers understand what type of apps are likely to attract more users.

```python
from csv import reader

### The Google Play data set ###
opened_file = open('googleplaystore.csv')
read_file = reader(opened_file)
android = list(read_file)
android_header = android[0]
android = android[1:]    #cut off head

### The App Store data set ###
opened_file = open('AppleStore.csv')
read_file = reader(opened_file)
ios = list(read_file)
ios_header = ios[0]
ios = ios[1:]   #cut off head

```

```python
    # loop through each slice and print number of rows/columns if exists

def explore_data(dataset, start, end, rows_and_columns=False):
        dataset_slice = dataset[start:end]    
        for row in dataset_slice:
            print(row)
            print('\n') # adds a new (empty) line after each row

        if rows_and_columns:
            print('Number of rows:', len(dataset))
            print('Number of columns:', len(dataset[0]))
```

```python
# checking for duplicate entries

duplicate_apps = []
unique_apps = []

for app in android:
    name = app[0]
    if name in unique_apps:
        duplicate_apps.append(name)
    else:
        unique_apps.append(name)

print(f"number of duplicate entires: {len(duplicate_apps)} \n")
print(f"examples of duplicate apps: {duplicate_apps[:10]}")
```

- will be removing duplicate entries and keeping only entires with highest rating number as it is more current data

```python
# remove duplicates
reviews_max = {}

for app in android:
    name = app[0]
    n_reviews = float(app[3])
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
    elif name not in reviews_max:
        reviews_max[name] = n_reviews
        
#print(len(reviews_max))
        
    
    
```

```python
# TODO use markdown to explain this step

android_clean = []
already_added = []

for app in android:
    name = app[0]
    n_reviews = float(app[3])
    if n_reviews == reviews_max[name] and name not in already_added:
        android_clean.append(app)
        already_added.append(name)
        


explore_data(android_clean, 0, 3, True)


    
```

```python
def is_english(string):
    non_ascii = 0
    
    for character in string:
        if ord(character) > 127:
            non_ascii += 1
    
    if non_ascii > 3:
        return False
    else:
        return True

#print(is_english('Docs To Goâ„¢ Free Office Suite'))
#print(is_english('Instachat ðŸ˜œ'))

```

```python
android_english = []
ios_english = []

for app in android_clean:
    name = app[0]
    if is_english(name):
        android_english.append(app)
        
for app in ios:
    name = app[1]
    if is_english(name):
        ios_english.append(app)
        
explore_data(android_english, 0, 3, True)
print('\n')
explore_data(ios_english, 0, 3, True)
```

```python
ios_english_free = []
android_english_free = []

for app in ios_english:
    if float(app[4]) == 0:
        ios_english_free.append(app)
        
for app in android_english:
    if (app[6]) == "Free":
        android_english_free.append(app)
        
explore_data(ios_english_free, 0, 3, True)


        
explore_data(android_english_free, 0, 3, True)
    
```

```python


def freq_table(dataset, index):
    table = {}
    total = 0
    
    for row in dataset:
        #counting total values in index
        total += 1
        #counting index values in column
        value = row[index]
        if value in table:
            table[value] += 1
        else:
            table[value] = 1
    
    #convert index values into total percent distribution
    table_percentages = {}
    for key in table:
        percentage = (table[key] / total) * 100
        table_percentages[key] = percentage 
    
    return table_percentages


def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    #convert dictionary pairs into list of tupples: (Key/column label: Value/column value)
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)
    #sorting list of tuples in descending order 
    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])


    
```

```python

display_table(ios_english_free, 1)

```

```python
display_table(android_english_free, 1)
```

- appstore does not have # of installs in data set. Will be using "user ratings" column as proxy. Not perfect, but I think this is the best course of action with the current dataset.

```python


genres_ios = freq_table(ios_english_free, -5)


for genre in genres_ios:
    total = 0
    len_genre = 0
    
    for app in ios_english_free:
        genre_app = app[-5]
            #match if app's genre is within current genre
        if genre_app == genre:        
            #get total number of ratings for app
            n_ratings = float(app[5])
            #accumulating ratings in genre
            total += n_ratings
            #count number of apps in genre
            len_genre += 1
            #calculate avg ratings/genre
    avg_n_ratings = total / len_genre
    print(genre, ':', avg_n_ratings)

```

```python
genres_android = freq_table(android_english_free, 1)

for genre in genres_android:
    
    total = 0
    len_category = 0
    for app in android_english_free:
        #compare genre of specificapp to app category
        category_app = app[1]
        if category_app == genre:
            #strip string of characters to convert to float
            n_installs = app[5]
            n_installs = n_installs.replace(',', '')
            n_installs = n_installs.replace('+', '')
            #convert striped string to float and sum to total
            total += float(n_installs)
            #add number of apps in category
            len_category += 1
            
    avg_n_installs = total / len_category
    print(genre, ':', avg_n_installs)
```
